```python
import pandas as pd
```

```python
df_0= pd.read_csv('account_values/account_value_20250521-18h02.csv')
df_0.drop(columns=['Unnamed: 0'], inplace=True)
df_0.set_index('date', inplace=True)
df_0
```

```python
df_2= pd.read_csv('account_values/account_value_20250521-18h03.csv')
df_2.drop(columns=['Unnamed: 0'], inplace=True)
df_2.set_index('date', inplace=True)
df_2
```

```python
df_0 = df_0.assign(
    ror = df_0['account_value'].pct_change(),
)
df_2 = df_2.assign(
    ror = df_2['account_value'].pct_change(),
)
```

```python
import matplotlib.pyplot as plt

palette = ['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E67E22']

def plot_account_value(df_list, column_name):
    plt.figure(figsize=(12, 6))
    for i in range(len(df_list)):
        plt.plot(df_list[i].index, df_list[i][column_name], label=f'df_{i} {column_name}', color=palette[i])
    plt.xlabel('Date')
    plt.ylabel(column_name)
    plt.title(f'{column_name} Comparison')
    plt.legend()
    plt.grid(True)
    plt.xticks([])  # Ignore x-tickers
    plt.show()
```

```python
plot_account_value([df_0, df_2], 'account_value')
```

```python
plot_account_value([df_0, df_2], 'ror')
```
